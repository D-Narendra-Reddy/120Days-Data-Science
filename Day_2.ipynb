# IF/ELSE  Examples

```python
# 1️ Write a program to check whether a number entered by the user is positive or negative.
Number=float(input("Enter a Number: "))

if(Number > 0):
  print("Number is Positive")
elif(Number < 0):
  print("Number is Negative")
else:
  print("Number is Zero")
```

```python
#2  Take an age as input and print:

# "Child" if age < 13

# "Teenager" if age is between 13 and 19

# "Adult" if age ≥ 20

Age=int(input("Enter Age: "))

if(Age<13):
  print("Child")
elif(Age>13 and Age<19):
  print("Teenager")
else:
  print("Adult")
```

```python
# 3 Accept two numbers and print the greater number. If both are equal, print "Both are equal"

x=float(input("Enter first number: "))
y=float(input("Enter second number: "))

if(x>y):
  print("First Number is Largest")
elif(y>x):
  print("Second Number is Largest")
elif(x==y):
  print("Both are equal")
else:
  print("Invalid Number")
```

```python
# 9 Check if a year is leap year or Not

year=int(input("Enter Year: "))

if(year%4==0) and (year%100!=0):
  print(year,"is Leap year")
else:
  print(year,"is not a Leap year")
```

```python
# 10 Check if a triangle is valid based on side lengths
a=int(input("Enter one side of triangle: "))
b=int(input("Enter second side of triangle: "))
c=int(input("Enter Third side of triangle: "))

if(a+b>c) and (b+c>a) and (c+a>b):
  print("Valid Triangle")
else:
  print("Not a Valid Triangle")
```

# For Loop Examples


```python
# 1 WAP to print numbers from 1 to 50 using a for loop
print("Numbers from 1 to 50")
for i in range(1,51):
  print(i)
```

```python
# 2 Take a number n and print the multiplication table of n using a for loop.
# (Example: if input = 5 → print 5×1, 5×2 … 5×10)

Number=int(input("Enter a Number: "))
for i in range(1,10+1):
  res=Number*i
  print(Number,"x",i,"=",res)
```

```python
# 3 Print the sum of all numbers from 1 to n using a for loop.
n=int(input("Enter nth number: "))
sum=0
for i in range(1,n+1):
  sum=sum+i
print("Sum of",n,"numbers =",sum)
```

```python
# 4 Print the square of each number from 1 to 10 using a for loop.
# (Example: 1 → 1, 2 → 4, 3 → 9, …)

for i in range(1,10+1):
  print(i**2)
```

# While Loop Examples


```python
# 1 Print numbers for 1 to 10 using while loop
i=1
while(i<=10):
  print(i)
  i+=1
```

```python
# 2 Reverse a number using while loop(eg: 123 --> 321)
x=int(input("Enter a number: "))
reverse=0
while(x>0):
  digit=x%10
  reverse=(reverse*10)+digit
  x=x//10
print(reverse)
```

```python
# 3 WAP to ask user repeatedly for input until they type "exit"
while True:
  x=input("Enter Text: ")
  if(x=="exit"):
    break
print("Program ended")
```

```python
# Prime Number Checker
number=int(input("Enter Number: "))
if(number<=1):
  print("Not a prime number")
else:
  for i in range(2,number):
    if(number%i==0):
      print("Not a prime number")
      break
  else:
    print("prime Number")
```

```python
num=int(input("Enter number: "))
y=int(num**0.5)+1
if(num<=1):
  print("Not Prime")
else:
  for i in range(2,y):
    if(num%i==0):
      print("Not a Prime")
  else:
    print("Prime Number")
```

```python

#prime number checking using optimized square root method using math import
import math as m
x = int(input("Enter the number:"))
y = int(m.sqrt(x))
if x <=1:
  print("Not a prime")
else:
  for i in range(2,y):
    if x%i==0:
      print("Not a prime")
      break
  else:
      print("Prime")

```

```python
# Factorial Of number
x=int(input("Enter Number: "))
fact=1

if(x==0):
  print("The Factorial of Number is 1 ")
elif(x==1):
  print("The Factorial of number is 1")
else:
  for i in range(1,x+1):
    fact=fact*i
  print("Factorial of number is",fact)
```

```python
# Fibonacci Series
x=int(input("Enter Numbers of terms you want: "))
first=0
second=1
print(first,second,end=" ")
for i in range(1,x+1):
  nxt=first+second
  print(nxt,end=" ")
  first=second
  second=nxt

```

```python
# Printing square pattern
x=int(input("Enter Number of rows: "))
for row in range(1,x+1):
  for column in range(1,x+1):
    print("*",end=" ")
  print()
```

```python
# Printing Right angled traingle
x=int(input("Enter Numbers of rows: "))
for row in range(1,x+1):
  print("* "*row)
```

```python
# Inverted Right Angle Triangle
x=int(input("Enter Numbers of rows"))
for row in range(x,0,-1):
  print("* "*row)
```

```python
# Pyramid Printing
x=int(input("Enter numbers of rows: "))

for row in range(1,x+1):
  for space in range(1,x-row+1):
    print(" ",end="")
  #printing stars
  for star in range(1,2*row):
    print("*",end="")

  print()
```

```python
# Inverted Pyramid

x=int(input("Enter Numbers of rows: "))
for row in range(x,0,-1):
  print(" "*(x-row),end="")
  print("*"*(2*row-1))
```

```python
# Numbers pyramid

x=int(input("Enter Number: "))
for i in range(1,x+1):
  for num in range(1,i+1):
    print(num,end=" ")
  print()
```
